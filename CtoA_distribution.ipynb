```python
with open('rpoB1_3new.csv', 'r') as f:
    data = f.readlines()

print(len(data))
bc_families=[]
for line in data:
    line=line.strip('\n')
    bc_family=line.split(',')
    if 'A' in bc_family:
        bc_families.append(bc_family)
        
    
```

```python
ROI="TGTCTCAGTTTATGGACCAGAACAACCCGCTGTCTGAGATTACGCACAAACGTCGTATCTCCGCACTCGGCCCAGGCGGTCT"

pos_lis=[]
for i in range(len(ROI)):
    if ROI[i]=='C':
        pos_lis.append(i)
    
```

```python
filtered_bc_families=[]
for pos in pos_lis:
    for family in bc_families:
        if family[pos]=='A':
            filtered_bc_families.append(family)
            
            
```

```python
len(filtered_bc_families)
```

```python
def count_As(lis):
    count=0
    for char in lis:
        if char=='A':
            count=count+1
    return count
```

```python
count_lis=[]
for family in filtered_bc_families:
    a_count=count_As(family)
    count_lis.append(a_count)
    
print(count_lis)
```
