```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#nodes rpoB1 -5345239
#nodes rpoB3 -3267731
def make_df(file1):

    df=pd.read_csv(file1,skiprows=2,delim_whitespace=True,header=None)
    df.columns=['pos','sub','count']
    filtered=df.loc[:,'sub'].str.contains("A|T|G|C")
    df_filtered=df.loc[filtered]
    df_filtered['sub']=df_filtered['sub'].apply(lambda x : x.split("->")[1])
    #df_filtered.loc[:,"sub"]=f
    #ROI="ACCTGTGGAGCTTTTTAAGTATGGCACGCGTAACTGTTCAGGACGCTGTAGAGAAAATTGGTAACCGTTTTGACCTGGTACTGG"
    ROI="TGTCTCAGTTTATGGACCAGAACAACCCGCTGTCTGAGATTACGCACAAACGTCGTATCTCCGCACTCGGCCCAGGCGGTCT"
    df_wide=df_filtered.pivot(index='pos', columns='sub', values='count')
    df_wide=df_wide.reindex(range(len(ROI)))
    df_wide=df_wide.fillna(value=0)
    ROI_lis=list(ROI)
    df_wide["roi"]=ROI_lis
    
    return df_wide

def filter_sub(df, nuc1,nuc2,sample):
    df_sub=df.loc[df['roi'] == nuc1]
    df_sub=df_sub.loc[:, nuc2:nuc2]
    df_sub=df_sub.rename(index=str, columns={nuc2:sample})
    
    return df_sub
```

```python
def make_df_lis(pad_lis,name_lis,fam_lis):
    df_lis=[]
    for i in range(len(pad_lis)):
        df_i=make_df("/home/britney/mds_files/rpoB"+pad_lis[i]+"_3")
        df_C_A=filter_sub(df_i,'C','A',name_lis[i])
        df_C_A=df_C_A/fam_lis[i]
        df_lis.append(df_C_A)
    merge= pd.concat(df_lis, axis=1)
    sums=pd.DataFrame(merge.sum(axis=0)) 
    #merge_sums=pd.concat([norm_0,norm_1,norm_2], axis=0)

    fig=sums.plot(kind='bar',figsize=(10,5),title="C->A Frequency",color="black",legend=None)
    
pad_lis=["07","52","24","16","61","33","25","71","42","34","06","51","43","15","62"]
fam_lis=[1410180,3748008,3040099,3932407,2932012,1476373,1278178,2291514,2817829,3121238,820462,1427296,1513127,2918034,2974947]
name_lis=['wt1','wt2','wt3','pca2yn1','pca2yn2','pca2yn3','mutT1','mutT2','mutT3','mutmy1','mutmy2','mutmy3','dinb1','dinb2','dinb3']
make_df_lis(pad_lis,name_lis,fam_lis)    
```
